```python
import pandas as pd
import gseapy as gp
import seaborn as sns
import matplotlib.pyplot as plt
import os
from fractions import Fraction
import seaborn as sns
import matplotlib.pyplot as plt

file_list = [
    "clusterProfiler_deg_phvted_vs_Cntrl.csv",
    "clusterProfiler_deg_HFpEF_dx_pvr_lesser_3_vs_PAH_dx_pvr_lesser_3_P_threshold_0.05.csv",
    "clusterProfiler_deg_HFpEF_dx_pvr_greater_3_vs_PAH_dx_pvr_lesser_3_P_threshold_0.05.csv",
    "clusterProfiler_deg_PAH_dx_pvr_greater_3_vs_phvted_dx_pvr_lesser_3_P_threshold_0.05.csv",
    "clusterProfiler_deg_HFpEF_dx_pvr_greater_3_vs_phvted_dx_pvr_greater_3_P_threshold_0.05.csv",
    "clusterProfiler_deg_Control_dx_pvr_lesser_3_vs_PAH_dx_pvr_greater_3_P_threshold_0.05.csv",
    "clusterProfiler_deg_PAH_dx_pvr_greater_3_vs_phvted_dx_pvr_greater_3_P_threshold_0.05.csv",
    "clusterProfiler_deg_HFpEF_dx_pvr_greater_3_vs_phvted_dx_pvr_lesser_3_P_threshold_0.05.csv",
    "clusterProfiler_deg_Control_dx_pvr_lesser_3_vs_HFpEF_dx_pvr_greater_3_P_threshold_0.05.csv",
    "clusterProfiler_deg_PAH_dx_pvr_greater_3_vs_PAH_dx_pvr_lesser_3_P_threshold_0.05.csv",
    "clusterProfiler_deg_Control_dx_pvr_lesser_3_vs_phvted_dx_pvr_lesser_3_P_threshold_0.05.csv",
]


for file in file_list:
    
    import seaborn as sns
    import matplotlib.pyplot as plt
    clust_data = pd.read_csv(file)
    df = clust_data
    
    # Only significant
    df = df[df['p.adjust'] < 0.05]

    # Sort by Enrichment Score (ES) 
    df['GeneRatio'] = df['GeneRatio'].apply(lambda x: float(Fraction(x)))
    df['BgRatio'] = df['BgRatio'].apply(lambda x: float(Fraction(x)))
    df['FoldEnrichment'] = df['GeneRatio'] / df['BgRatio']
    df['Description'] = df['Description'].astype(str).str.replace(r'[\[\]]', '', regex=True)
    df_sorted = df.sort_values(by="FoldEnrichment", ascending=False)
    df_sorted['Description'] = df_sorted['Description'].apply(lambda x: x.replace('[', '').replace(']', '') if isinstance(x, str) else x)

    df_top10 = df_sorted.head(20)
    plt.figure(figsize=(10, 6))
    sns.barplot(x="FoldEnrichment", y="Description", data=df_top10, palette="viridis")

    plt.xlabel('FoldEnrichment')
    plt.ylabel('Description')
    plt.title(f'Top 20 Significant Enrichment Scores for GO:BP for file:{file}')

    plt.tight_layout()
    plt.show()  # Display the plot for the current file
    
    del plt
    del sns


```
